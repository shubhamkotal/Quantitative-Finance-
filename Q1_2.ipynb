```python
import pandas as pd
import numpy as np

data = pd.read_csv('MScFE 650 MLF GWP Data.csv', index_col='Date')
data.head()
```

```python
from scipy import stats

#calculate 0% trimmed mean
mean = stats.trim_mean(data, 0)

#calculate 5% trimmed mean
mean_5p = stats.trim_mean(data, 0.05)

#calculate 50% trimmed mean
mean_50p = stats.trim_mean(data, 0.50)

print("Array of means of data set is: \n" + str(mean))
print("Array of means of 5% trimmed data set is: \n" + str(mean_5p))
print(data.describe())
```

```python
# Assigning random weights of 0.3, 0.4 and 0.3 to each mean
# Choosing MSCI Singapore as consideration factor

Median_returns = 0.3*data['MSCI SINGAPORE'].describe(percentiles=[0.5])
Mean_returns = 0.3*data['MSCI SINGAPORE'].mean()
Mean_5p_returns = 0.4*(stats.trim_mean(data['MSCI SINGAPORE'], 0.05))
```

```python
print('Returns on 5% trim mean series is :' + str(Mean_5p_returns))
print('Returns on Median series is :' + str(Median_returns['50%']))
print('Returns on Mean series is :' + str(Mean_returns))
```
