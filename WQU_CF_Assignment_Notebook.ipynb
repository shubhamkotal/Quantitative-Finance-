```python
# Let's start by putting the following (completely fictional) share price lists into NumPy arrays

microsoft = [120.3, 121.7, 119.8, 122.5, 123.3]
apple = [124.1, 124.3, 124.2, 124.8, 124.3]
facebook = [110.2, 112.4, 116.8, 115.3, 119.6]

# TASK 1: Import NumPy and create arrays
import numpy as np

microsoft_array = np.array(microsoft)
apple_array = np.array(apple)
facebook_array = np.array(facebook)

print("Microsoft's numpy array: ", type(microsoft_array))
print(microsoft_array)

print("Apple's numpy array: ",type(apple_array))
print(apple_array)

print("Facebook's numpy array: ",type(facebook_array))
print(facebook_array)
```

```python
# Now let's use some NumPy functions to learn some interesting statistical information


# TASK 2A: Calculate the mean share price for each company, and store in suitably named variables.
#          Then print them out (in order - Microsoft, then Apple, then Facebook).

microsoft_mean = microsoft_array.mean()
apple_mean = apple_array.mean()
facebook_mean = facebook_array.mean()

print("Mean values:")
print("Microsoft's mean: ", microsoft_mean)
print("Apple's mean: ", apple_mean)
print("Facebook's mean: ", facebook_mean)

print('-'*50)

# TASK 2B: Calculate the standard deviation of share prices for each company using a NumPy function,
#          and store in suitably named variables. Then print them out (in order - Microsoft, then Apple, then Facebook).
microsoft_std = microsoft_array.std()
apple_std = apple_array.std()
facebook_std = facebook_array.std()

print("Std. deviation values:")
print("Microsoft's std: ", microsoft_std)
print("Apple's std: ", apple_std)
print("Facebook's std: ", facebook_std)

print('-'*50)

# TASK 2C: Calculate the percentage change in share price from the first price to the last price, and store 
#          in suitably named variables. Then print them out (in order - Microsoft, then Apple, then Facebook).

# There was a little ambiguity in if the percentage change has to be between start and end prices or change in percentage between 2 prices
# Hence I have calculated both

# Percent change between first and last price
def startEndPercentChange( np_array ):
    """Calculates Percent change between first and last price """
    start = np_array[0]
    end = np_array[1]
    
    return  ((end - start)/start)*100

pc_microsoft = startEndPercentChange(microsoft_array)
pc_apple = startEndPercentChange(apple_array)
pc_facebook = startEndPercentChange(facebook_array)

print("% change values:")
print("Microsoft's Percent Change: ", pc_microsoft)
print("Apple's Percent Change: ", pc_apple)
print("Facebook's Percent Change: ", pc_facebook)

print('*'*50)

def calculatePercentChange( np_array ):
    """ Calculates and returns percentage change for a one dimensional array"""
    pct_change = np.diff(np_array) / np_array[1:]
    return pct_change * 100

microsoft_pc = calculatePercentChange(microsoft_array)
apple_pc = calculatePercentChange(apple_array)
facebook_pc = calculatePercentChange(facebook_array)

print("% change values:")
print("Microsoft's Percent Change Array: ", microsoft_pc)
print("Apple's Percent Change Array: ", apple_pc)
print("Facebook's Percent Change Array: ", facebook_pc)
```

```python
# Now we're going to load our NumPy arrays into a Pandas Dataframe
# The 5 values in each company's array is the share price for the first 5 months of 2018

months = ["Jan 2018", "Feb 2018", "Mar 2018", "Apr 2018", "May 2018"]
company_names = ["Microsoft", "Apple", "Facebook"]
prices = [microsoft_array, apple_array, facebook_array]

# TASK 3: Import Pandas and load the arrays into a dataframe with appropriate row and column labels, and then
#         output the dataframe by simply typing its variable name on a new line

import pandas as pd

stock_prices = pd.DataFrame(prices, index=company_names, columns = months)

# Print the dataframe
stock_prices
```

```python
# Finally, let's draw some interesting graphs! 
# You will need to use some of the statistical values you calculated earlier in order to do this

# TO DO: Import Matplotlib's Pyplot submodule
import matplotlib.pyplot as plt

# TASK 4A: Draw a line graph for one of the companies, showing their share prices over time, using the dataframe.
#          Provide suitable labels. Hint: using your_dataframe_variable.loc["your_row_label", :] will extract a single row.
plt.xlabel("Time")
plt.ylabel("Stock Prices")
plt.title("Line graph for Microsoft's prices")
plt.plot(stock_prices.loc["Microsoft", :])
plt.show()

plt.xlabel("Time")
plt.ylabel("Stock Prices")
plt.title("Line graph for Apple's prices")
plt.plot(stock_prices.loc["Apple", :])
plt.show()

plt.xlabel("Time")
plt.ylabel("Stock Prices")
plt.title("Line graph for Facebook's prices")
plt.plot(stock_prices.loc["Facebook", :])
plt.show()

# TASK 4B: Draw a bar chart, using any data structure, comparing each company's value for any statistic of your choice 
#          from Task 2 (A, B or C)

mean_data = [microsoft_mean, apple_mean, facebook_mean]
plt.xlabel("Company Names")
plt.ylabel("Mean")
plt.title("Bar chart for mean")
plt.bar(company_names, mean_data, alpha = 0.5)

plt.show()


std_data = [microsoft_std, apple_std, facebook_std]
plt.xlabel("Company Names")
plt.ylabel("Standar Deviation")
plt.title("Bar chart for std")
plt.bar(company_names, std_data, alpha = 0.5)

plt.show()

pc_data = [pc_microsoft, pc_apple, pc_facebook]
plt.xlabel("Company Names")
plt.ylabel("Percent Change between Jan 2018 to May 2018")
plt.title("Bar chart for percent change")
plt.bar(company_names, pc_data, alpha = 0.5)

plt.show()
```
