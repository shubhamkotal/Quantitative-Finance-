#### Question 1 : Financial Products

#### 1.1 - What is an Exchange-Traded Fund (ETF)?
An exchange-traded fund (ETF) is a form of investment that owns several underlying assets rather than just one, like a stock does. It is a form of investment that follows an index, sector, commodity, or other asset and may be bought and sold on a stock market much like a regular stock. An ETF may be set up to track anything from a single commodity's price to a vast and varied group of assets.Because it is exchanged on an exchange like stocks, an ETF is termed an exchange traded fund. As shares are purchased and sold on the market, the price of an ETF's shares will fluctuate during the trading day. [Investopedia](https://www.investopedia.com/terms/e/etf.asp)

Varios Types of ETFs are:

- Bond ETFs 
- Stock ETFs
- Currency ETFs
- Commodity ETFs
- Inverse ETFs
- Industry ETFs

#### 7.2 Show the amount of variation Explained by 5 component

```python
pca = PCA(n_components= 5)
pca.fit(df)
X1=pca.fit_transform(df)

print(pca.explained_variance_ratio_) 
```
