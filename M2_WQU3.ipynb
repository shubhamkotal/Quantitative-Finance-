Q5.

```
t = np.array([2.0, 3.0, 5.0, 7.0,10.0,20.0,30.0])

for index, row in US.iterrows():
    y=np.array(row)
    curve, status = calibrate_ns_ols(t, y, tau0=1.0)  # starting value of 1.0 for the optimization of tau
    s = np.linspace(0, 20, 100)
    y=curve
    plot(s,y(s),marker='o')
    plt.title(index)
    plt.xlabel('Year')
    plt.ylabel('US Treasury yields')
    plt.show()

print('Nelson-siegel Curve for each day from month october and November')
```

```
oct=np.array(US[0:23].mean())
nov=np.array(US[23:].mean())

t = np.array([2.0, 3.0, 5.0, 7.0,10.0,20.0,30.0])
y_oct=np.array(oct)
y_nov=np.array(nov)

curve_oct, status = calibrate_ns_ols(t, y_oct, tau0=1.0)  # starting value of 1.0 for the optimization of tau
curve_nov, status = calibrate_ns_ols(t, y_nov, tau0=1.0)  # starting value of 1.0 for the optimization of tau
from matplotlib.legend_handler import HandlerLine2D

s = np.linspace(0, 20, 100)
line1,=plot(s,curve_oct(s),marker='o',color='red',label='October')
line2,=plot(s,curve_nov(s),marker='o',color='green',label='November')
plt.legend(handler_map={line1: HandlerLine2D(numpoints=4)})
plt.title("MEAN YIELD FOR OCTOBER AND NOVEMBER")
plt.xlabel('Year')
plt.ylabel('US Treasury yields')
plt.show()
```

```
GLD_mod_oct = ARMA(GLD['Adj Close'][0:23], order=(1, 1))
GLD_res_oct = GLD_mod_oct.fit()
GLD_mod_nov = ARMA(GLD['Adj Close'][23:], order=(1, 1))
GLD_res_nov = GLD_mod_nov.fit()
INDA_mod_oct = ARMA(INDA['Adj Close'][0:23], order=(1, 1))
INDA_res_oct = INDA_mod_oct.fit()
INDA_mod_nov = ARMA(INDA['Adj Close'][23:], order=(1, 1))
INDA_res_nov = INDA_mod_nov.fit()

# Print out summary information on the fit
print(GLD_res_oct.summary())
print(GLD_res_nov.summary())
print(INDA_res_oct.summary())
print(INDA_res_nov.summary())
```

```
import statsmodels.api as sm
sm.graphics.tsa.plot_acf(GLD['Adj Close'][0:23], lags=10)
sm.graphics.tsa.plot_pacf(GLD['Adj Close'][0:23], lags=10)
sm.graphics.tsa.plot_acf(GLD['Adj Close'][23:], lags=10)
sm.graphics.tsa.plot_pacf(GLD['Adj Close'][23:], lags=10)
plt.show()
```

```
from pylab import rcParams
import statsmodels.api as sm
import matplotlib.pyplot as plt

#decomposing for more better diagnostic purpose
rcParams['figure.figsize'] = 18, 8
decomposition1 = sm.tsa.seasonal_decompose(GLD['Adj Close'][0:-1], model='additive')
decomposition3 = sm.tsa.seasonal_decompose(INDA['Adj Close'][0:-1], model='additive')
fig1 = decomposition1.plot()
fig3 = decomposition3.plot()
plt.show()
```

Lower the AIC value, Better the fit. Hence INDA (Equity ETF) ARMA model for november month is the best one.
Looking after decomposition charts its clear that both ETF have seasonality effects and noise present.

```
import numpy as np
returns_GLD_oct = pd.DataFrame(np.diff(np.log(GLD['Adj Close'][0:23].values)))
returns_GLD_oct.index = GLD[0:23].index.values[1:GLD[0:23].index.values.shape[0]]
returns_GLD_oct.columns = ['GLD OCT Returns']

returns_GLD_nov = pd.DataFrame(np.diff(np.log(GLD['Adj Close'][23:].values)))
returns_GLD_nov.index = GLD[23:].index.values[1:GLD[23:].index.values.shape[0]]
returns_GLD_nov.columns = ['GLD NOV Returns']

returns_INDA_oct = pd.DataFrame(np.diff(np.log(INDA['Adj Close'][0:23].values)))
returns_INDA_oct.index = INDA[0:23].index.values[1:INDA[0:23].index.values.shape[0]]
returns_INDA_oct.columns = ['INDA OCT Returns']

returns_INDA_nov = pd.DataFrame(np.diff(np.log(INDA['Adj Close'][23:].values)))
returns_INDA_nov.index = INDA[23:].index.values[1:INDA[23:].index.values.shape[0]]
returns_INDA_nov.columns = ['INDA NOV Returns']

plt.figure(figsize=(15,5));
plt.plot(returns_GLD_oct.index,returns_GLD_oct,color='red');
plt.plot(returns_GLD_nov.index,returns_GLD_nov,color='blue');
plt.plot(returns_INDA_oct.index,returns_INDA_oct,color='green');
plt.plot(returns_INDA_nov.index,returns_INDA_nov,color='orange');
plt.ylabel('Returns');
plt.title('GLD Returns');
```

Red- returns GLD oct

Green- returns INDA oct

Blue- returns GLD nov

Orange- returns INDA nov
